## 4.1 Variabel string

Strings adalah serangkaian karakter yang tersusun secara berurutan. Karakter tersebut bermakna bahwa apa yang disimpan di dalam Strings dapat berupa huruf `'a', 'b', 'c'`, atau angka `'1', '2', '3' `, atau karakter khusus seperti `'#', '$', '%', '[' `. Tersusun secara berurutan maksudnya adalah setiap karakter di dalam Strings memiliki nomor posisi atau indeks.

````python
teks = 'karakcer'
print(teks)
print(teks[5])
````

variabel teks berisi nilai 'karakcer'. Sesuai definisi Strings, bahwa string adalah serangkaian karakter, maka variabel string teks berisi `'k', 'a', 'r', 'a', 'k', 'c', 'e', 'r' `. Dengan demikian, variabel teks bisa di `print() ` secara utuh atau ditampilkan hanya sebuah karakter saja berdasarkan posisi indeksnya. Contoh di atas ditampilkan sebuah karakter pada posisi indeks ke-5 yang nilai datanya berupa `' c '`.

Perlu dicatat bahwa Strings bersifat *immutable* atau tidak dapat diubah. Makna *immutable* di sini adalah **sebuah karakter di dalam string** tersebut **tidak dapat diubah**, kecuali mengganti **nilai seluruh variabel string**. Perhatikan ilustrasi berikut ini.

Apabila ingin mengganti satu karakter yang salah tersebut, tidak dapat dilakukan dengan mengubah berdasarkan indeksnya.

````python
teks[5] = 't'
````

Yang dapat dilakukan hanyalah mengubah seluruh nilai variabel tersebut seperti contoh berikut

````python
teks = 'karakter'
````

Mengubah nilai seluruh isi teks tersebut sebenarnya juga membentuk sebuah objek baru di dalam memori komputer. Maksudnya bagaimana ? Nah coba baca [literatur ini](https://medium.datadriveninvestor.com/mutable-and-immutable-python-2093deeac8d9) tentang **id memory** suatu objek di dalam memory komputer.

````python
teks_1 = 'karakcer'
print(id(teks_1))
teks_1 = 'karakter'
print(id(teks_1))
````

Dengan demikian, meskipun nama variabel masih sama yakni `teks_1`, sebenarnya di dalam memori komputer, variabel string tesebut adalah 2 objek yang berbeda.

## 4.3 Operasi-Operasi Strings
### 4.3.1 Penggabungan 'Concatenation'

Dua buah string dapat digabungkan menggunakan operator ` + `. Perhatikan contoh berikut ini. Penggunaan string yang berisi spasi `' '` dapat dimanfaatkan untuk menghasilkan nama_lengkap yang baik.

````python
nama_depan = 'Erny'
nama_tengah = 'Qurotul'
nama_belakang = 'Ainy'
nama_lengkap = nama_depan + ' ' + nama_tengah + ' ' + nama_belakang

print(nama_lengkap)
````

### 4.3.2 Panjang 'len' String

Kadang kala diperlukan informasi panjang karakter suuatu string. Fungsi `len()`  digunakan untuk menghitung panjang parakter sebuah variabel string.

````python
nama_depan = 'Erny'
nama_tengah = 'Qurotul'
nama_belakang = 'Ainy'
print(len(nama_depan))
print(len(nama_tengah))
````

### 4.3.3 Mengakses Karakter

Mengakses sebuah karakter dapat langsung menuliskan nomor indeksnya.

````python
teks_3 = 'Lagu Puitis Tulus'
print('isi indeks ke-0 adalah ',teks_3[0])
print('isi indeks ke-7 adalah ',teks_3[7])
````

Mengakses sekumpulan karakter dapat menggunakan `[ : ]`. **Ingat!** bahwa posisi (indeks) karakter berawal dari ` 0 `.  Perhatikan contoh di bawah ini

````python
teks_3 = 'Lagu Puitis Tulus'
print(teks_3[1:5])  # dari posisi (indeks) 1 karakter ke 5
print(teks_3[:5])   # dari awal sampai karakter ke 5
print(teks_3[2:])   # dari posisi (indeks) 2 sampai  karakter terakhir
````

````python
bulan = 'JanFebMarAprMayJunJulAugSepOctNovDec'
print(bulan[:3])    # dari indeks awal sampai karakter ke-3
print(bulan[0:3])   # dari posisi (indeks) 0 sampai karakter  ke- 3
print(bulan[3:6])   # dari indeks ke-3 karakter ke- 6
````

perhatikan implementasi berikut ini

````python
kbulan = int(input("Tampilkan bulan ke (1 s.d. 12): "))
indeks_awal = (kbulan -1 ) * 3
karakterke = indeks_awal + 3
print(bulan)
print(indeks_awal)
print(karakterke)
print(bulan[indeks_awal:karakterke])
````

### 4.3.5 Memecah String

Memecah string dapat dilakukan menggunakan fungsi `split()`. Perhatikan contoh

````python
pesan = "Salam hormat untuk semua, mereka yang berjasa membentuk kepribadianku."
print(pesan.split(' '))
print(pesan.split(','))
````

### 4.3.6 Menghitung Jumlah Karakter

Fungsi `count()` akan menghasilkan hitungan jumlah karakter yang dicari. Perhatikan contoh

````python
pesan = "Salam hormat untuk semua, mereka yang berjasa membentuk kepribadianku."
print('jumlah spasi adalah ', pesan.count(' ')) # hitung jumlah spasi
print('jumlah karakter "a" adalah ', pesan.count('a')) # hitung jumlah karakter a
````

### 4.3.7 Mengganti Karakter

Fungsi `replace()` akan mengganti suatu karakter tertentu. Penggantian ini sifatnya tidak permanen. Perhatikan contoh

````python
pesan = "Salam hormat untuk semua, mereka yang berjasa membentuk kepribadianku."
print('cek id di memori : ', id(pesan))
print(pesan.replace('u','oe'))
print('cek id di memori : ', id(pesan))
print(pesan) # teks pesan masih sama

````

### 4.3.8 Mencari suatu Karakter

Fungsi `find()` akan mencari suatu karakter tertentu. Jika ditemukan, akan ditampilkan posisinya (indeks). Jika tidak ditemukan akan menampilkan nomor indeks -1. Jika yang dicari adalah sekumpulan karakter, misal kata, maka yang ditampilkan adalah indeks awal dari kata tersebut. Jika yang dicari hanyalah sebuah karakter, maka indeks yang ditampilkan adalah indeks yang pertama saja. Perhatikan contoh

````python
pesan = "Salam hormat untuk semua, mereka yang berjasa membentuk kepribadianku."
print('cari kata hormat : ', pesan.find('hormat'))
print('cari karakter u : ', pesan.find('u'))
````

### 4.3.9 Mengkonversi Tipe Data Lain ke Strings

Fungsi `str()` akan mengkonversi tipe data lain menjadi berbentuk string. Perhatikan contoh

````python
kode_rahasia = "twob"
print(kode_rahasia)
kode_rahasia = kode_rahasia + str(2) + 'b'
print(kode_rahasia)

````

### 4.3.10 Membandingkan Strings

Membandingkan sebuah string dengan string lain dapat menggunakan operator ` == ` untuk kesamaan, dan untuk ketidaksamaan menggunakan ` != ` . Perhatikan contoh

````python
kode_rahasia = 'twob2b'
kode_pembuka = '2b2b'
print(kode_rahasia == kode_pembuka)
kode_pembuka = 'twob2B'
print(kode_rahasia == kode_pembuka)
kode_pembuka = 'twob2b'
print(kode_rahasia == kode_pembuka)
print(kode_pembuka != 'twob2b')
````

### 4.3.11 Memformat Strings

Memformat sebuah string menggunakan funsgi `.format()`. Perhatikan beberapa contoh kombinasi penggunaan format string.

````python
salam = "Selamat pagi {}"
print(salam.format("Pak Guru"))

salam2 = 'Selamat pagi {} dan {}'
pihak1 = 'Pak Guru'
pihak2 = 'Bu Guru'
print(salam2.format(pihak1, pihak2))

salam3 = 'Selamat {2} {0} dan {1}'
print(salam3.format(pihak1, pihak2, 'sore'))

salam4 = 'Selamat {waktu} {pejabat1} dan {pejabat2}'
print(salam4.format(pejabat1='Bupati Karanganyar', pejabat2='Kepala Sekolah', waktu='sore'))
````

Salah satu penggunaan `format()` yang menarik adalah membuat format string dengan lebar kolom dan alignment tabel sederhana menggunakan string. Penggunaan `{:25}` bermakna memberikan ruang sebanyak 25 karakter yang dapat diisi oleh suatu string.

````python
print('|{:25}|{:25}|'.format('kolom-1 lebar 25  ', 'Kolom 2'))
````

Pengaturan alignment teks menggunakan beberapa tambahan simbol sebagai berikut.

* ` < ` rataan kiri (format dasarnya),
* ` > ` rataan kanan,
* ` ^ ` rata tengah.

````python
print('|{:<25}|{:^25}|{:>25}|'.format('kolom-1 rata kiri  ', 'Kolom 2 Rata Tengah', 'Kolom 3 Rata Kanan', 'Kolom 2 Rata Kanan'))
````

Penggunaan format juga dapat digunakan memberikan format angka dalam ribuan dengan separator koma atau titik (untuk kasus di Indonesia).

````python
print('{:,}'.format(1234567890))
print('{:,}'.format(1234567890.0))

#Modifikas dengan separator menggunakan tanda .
print('{:.4f}'.format(346578987654323.141592653589793))
````
